```python
import os
import shutil
import pandas as pd

def copy_images(source_folder, destination_folder, image_names):
    for image_name in image_names:
        source_file = os.path.join(source_folder, image_name)
        destination_file = os.path.join(destination_folder, image_name)

        if os.path.isfile(source_file):
            shutil.copy2(source_file, destination_file)
            print(f"Image '{image_name}' has been copied.")
        else:
            print(f"Image '{image_name}' not found in the source folder.")
```

```python
df = pd.read_csv("train and test ids.csv")
```

```python
def ids_to_file_list(ids_list,file_initial,file_format):
    file_list = []
    for ids in ids_list:
        str_id = str(ids)
        if len(str_id) == 1:
            str_id = "00"+str_id
        elif len(str_id) ==2:
            str_id ="0"+str_id
        file_name = str(file_initial)+str_id+str(file_format)
        file_list.append(file_name)
    return(file_list)  
```

```python
test_ids_list = list(df["test_ids"].dropna().astype(int))
train_ids_list = list(df["train_ids"].dropna().astype(int))
```

```python
train_image_ids = ids_to_file_list(train_ids_list,"crop_row_",".jpg")
train_label_ids = ids_to_file_list(train_ids_list,"crop_row_",".npy")
test_image_ids = ids_to_file_list(test_ids_list,"crop_row_",".jpg")
```

```python
copy_images('Original_Data/Images/', 'Data/train/images/', train_image_ids)
copy_images('Original_Data/Labels/', 'Data/train/labels/', train_label_ids)
copy_images('Original_Data/Images/', 'Data/test/images/', test_image_ids)
```
